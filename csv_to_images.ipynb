```
data_dir = 'Data'
data_fear = 'Data/Fear/'
data_happy = 'Data/Happy/'
data_sad = 'Data/Sad/'

if(os.path.isdir(data_dir) == False):
  os.mkdir(data_dir)

if(os.path.isdir(data_fear) == False):
  os.mkdir(data_fear)

if(os.path.isdir(data_happy) == False):
  os.mkdir(data_happy)

if(os.path.isdir(data_sad) == False):
  os.mkdir(data_sad)
```

```
data = pd.read_csv(csv_filename)
df = pd.DataFrame(data)
df.head()
```

```
num_of_instances = df.shape[0]

for i in range(num_of_instances):
  img = df.iloc[i, 1:].values.tolist()
  pixels = np.array(img, 'uint8')
  emotion = df.iloc[i, 0].lower()
  image = pixels.reshape(48,48)

  if(emotion == 'fear'):
    image_path = data_fear
  if(emotion == 'sad'):
    image_path = data_sad
  if(emotion == 'happy'):
    image_path = data_happy
  
  path = image_path+f"{i}_{emotion}.jpg"
  cv2.imwrite(path, image)
```
